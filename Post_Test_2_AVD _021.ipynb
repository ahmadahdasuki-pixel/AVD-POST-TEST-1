### **Outliers Values**

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

Dari output di atas dapat diketahui bahwa ada beberapa kolom yang memiliki nilai ekstrem atau tidak sesuai dengan kriteria sebelumnya.

1. Clicks: Dengan Presentase 0,5%

2. Conversions:Dengan Presentase 4,3%

3. Clicks (Duplicate):Dengan Presentase 0%

## **Eksplorasi Data Analysis**

### **Comparrison/Perbandingan**

```python
sesuaikan = df.groupby('Channel')['Clicks'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x=sesuaikan.index, y=sesuaikan.values, palette='viridis', hue=sesuaikan.index, legend=False)
plt.title('PERBANDINGAN')
plt.xlabel('Channel')
plt.ylabel('Clicks')
plt.xticks(rotation=45)
plt.show()
```

Aktivitas:Menampilkan Perbandingan jumlah Clicks berdasarkan Channel Kampanye

Tujuan:Untuk Mengetahui Channel mana yang masih dan paling efektif dan mengetahi channel mana yang menarik pelanggan

Visualisasi:Bar Chart(Diagram Batang)

### **Composisi/Komposisi**

```python
Clicks_by_Channel = df.groupby('Channel')['Clicks'].sum().sort_values(ascending=False)
Clicks_by_Channel_top3 = Clicks_by_Channel.head(3)

plt.figure(figsize=(10, 8))
Clicks_by_Channel_top3.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Proporsi 3 Kategori Saluran Teratas Berdasarkan Hasil Kampanye ')
plt.ylabel('')
plt.axis('equal')
plt.show()
```

Aktivitas:Menampilkan proporsi 3 channel pemasaran teratas berdasarkan total jumlah klik yang dihasilkan dari kampanye.

Tujuan:Untuk Mengetahui channel mana yang mendominsi teratas dan memudahkan dalam interpretasi  strategi Marketing

Visualisasi:Pie Chart(Grafik Lingkaran)

### **Relationship/Hubungan**

```python
plt.figure(figsize=(8, 6))
sns.heatmap(data=df[['Impressions', 'Clicks']].corr(),
            annot=True,
            cmap='viridis',
            fmt='.2f')
plt.title('Korelasi antara Impressions dan Clicks')
plt.show()
```

Aktivitas:Menampilkan hubungan korelasi antara dua variabel penting dalam kampanye pemasaran, yaitu Impressions dan Clicks

Tujuan:Untuk mengetahui hubungan antar jangkauan dan interaksi yang dimana bisa di lihat juga bahwa hubungan antara impressions dan clicks itu kuat dan posistif karena nilai mereka sama sama naik
