### **Outliers Values**

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

Dari output di atas dapat diketahui bahwa ada beberapa kolom yang memiliki nilai ekstrem atau tidak sesuai dengan kriteria sebelumnya.

Dari Output Boxplot di atas dapat di ketahui bahwa data tersebut perlu di lakukan imputasi atau treatment di karenakan ada satu titik sangat jauh yaitu di bawah -1000,karena tidak mungkin orang mengeluarkan biaya itu di bawah -1000

```python
columns_to_impute = ["Spend"]

for col in columns_to_impute:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Menggunakan .loc[] agar tidak muncul SettingWithCopyWarning
    df.loc[:, col] = df[col].clip(lower=lower_bound, upper=upper_bound)
```

## **Boxplot**

```python
# Boxplot (Matplotlib): Penjualan berdasarkan Top 5 Negara

# Identifikasi 5 negara teratas berdasarkan jumlah entri
top5negara = df['Channel'].value_counts().head(5).index

# Filter DataFrame untuk hanya menyertakan 5 negara teratas
df_top5_countries = df[df['Channel'].isin(top5negara)]

plt.figure(figsize=(12, 7)) # Ukuran figure
df_top5_countries.boxplot(column='Clicks', by='Channel', ax=plt.gca(), grid=False, patch_artist=True, medianprops=dict(color='red'))
plt.title('Distribusi Saluran Yang Paling Banyak Di Klik', fontsize=16) # Judul
plt.xlabel('Saluran', fontsize=12) # X Keterangan
plt.ylabel('Jumlah Klik', fontsize=12) # Y Keterangan
plt.xticks(rotation=45, ha='right')
plt.suptitle('') # Menghilangkan judul default dari boxplot.by
plt.tight_layout()
plt.show()
```

Dari Visualisasi di atas dapat di ketahui:

Email memiliki jumlah kli di kisaran 500-2000 dengan nilai maksimumnya di atas 4000 namun terdapat outlier disini hampir mencapai 5000 yang berarti performa email itu tidak selalu konsisten karena bisa menghasilkan lonjakan klik yang lebih besar dari kebanyakan kampanye lainnya.

Facebook memiliki potensi dan performa kampanye yang cukup tinggi karena nilai maksimumnya mencapai kisaran 4500 yang berarti kampanye saluran facebook bisa lebih besar lagi.

Instagram dan Google Ads lebih stabil, box pada Instagram dan Google Ads terlihat cukup seimbang tanpa terlalu banyak outlier.

TikTok memiliki banyak outlier tinggi, TikTok terlihat beberapa titik outlier di atas kisaran 4500 klik.Ini menunjukkan bahwa ada beberapa kampanye melalui saluran TikTok yang performanya jauh lebih tinggi dari rata-rata atau nilai maksimum.

Action Yang dapat di lakukan:

## **Heatmap**

```python
plt.figure(figsize=(8, 6))
sns.heatmap(data=df[['Spend', 'Clicks', 'Conversions', 'Impressions']].corr(),
            annot=True,
            cmap='viridis',
            fmt='.2f')
plt.title('Korelasi antara Spend, Clicks, Conversions, Impressions')
plt.show()
```

Dari Visualisasi di atas dapat di ketahui bahwa:

Clicks memiliki korelasi paling kuat dengan Conversions (0.81),karena nilai korelasi antara Clicks dan Conversions = 0.81, yang merupakan salah satu nilai tertinggi pada heatmap yang dimana artinya semakin banyak klik yang diperoleh, semakin besar kemungkinan terjadi konversi.

Clicks juga memiliki korelasi kuat dengan Impressions (0.80)karena nilai korelasi 0.80 menunjukkan hubungan kuat antara Impressions dan Clicks yang di mana mendekati angka positif 1 yang berarti semakin banyak iklan ditampilkan, semakin besar kemungkinan pengguna mengklik iklan tersebut.

Spend memiliki korelasi kuat dengan Clicks (0.79),karena nilai korelasi antara Spend dan Clicks = 0.79 yang berarti semakin besar budget marketing yang dikeluarkan, semakin banyak klik yang diperoleh.

Korelasi Spend dengan Conversions cukup kuat (0.66),karena nilai 0.66 menunjukkan hubungan positif tetapi tidak terlalu kuat artinya konversi tidak bergantung pada besarnya budget atau biaya yang di keluarkan.

Korelasi Impressions dengan Conversions moderat (0.65),karena nilai 0.65 ini menunjukkan bahwa impressions tidak selalu langsung menghasilkan konversi yang dimana banyak orang melihat iklan, tetapi tidak semuanya melakukan pembelian atau tindakan.
